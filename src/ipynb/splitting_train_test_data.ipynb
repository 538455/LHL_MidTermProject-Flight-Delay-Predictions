```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
flights_sample = pd.read_csv('../../data/raw/200K_random_flights(26Nov).csv')
flights_sample.head(2)
```

```python
flights_train, flights_test = train_test_split(flights_sample, test_size=0.2, random_state=42)
```

```python
flights_train = flights_train.reset_index().drop('index', axis=1)
flights_train.head(2)
```

```python
flights_test = flights_train.reset_index().drop('index', axis=1)
flights_test.head(2)
```

```python
flights_train.to_csv('../../data/raw/flights_train.csv', index=False)
flights_test.to_csv('../../data/raw/flights_test.csv',index=False)
```
